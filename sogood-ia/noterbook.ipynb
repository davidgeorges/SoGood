```python
# 1. Importer les librairies
import pandas as pd
```

```python
# 2. Charger le fichier (adapte le chemin et le séparateur selon ton fichier : ',' pour CSV, '\t' pour TSV)
# Utilise la variable 'file' pour le chemin du fichier
df = pd.read_csv('./data.csv', low_memory=False, sep='\t', nrows=1000, on_bad_lines='skip')  # ajoute sep="," ou sep="\t" si besoin
```

```python
# 3. Nombre de lignes et colonnes
print("Shape :", df.shape)
```

```python
# 4. Afficher les premières lignes du jeu de données
display(df.head())
```

```python
# 5. Lister toutes les colonnes
print("Colonnes :", df.columns.tolist())
```

```python
# 6. Afficher les types de données par colonne
print("Types de colonnes :")
print(df.dtypes)
```

```python
# 7. Vérifier le taux de valeurs manquantes pour chaque colonne (en %)
missing = df.isnull().mean().sort_values(ascending=False) * 100
print("Taux de valeurs manquantes (%) :")
print(missing)
```

```python
# 8. Sélectionner les colonnes d'intérêt (modifie si besoin)
cols_interet = ['product_name', 'brands', 'categories', 'ingredients_text', 
                'nutriscore_score', 'nutriscore_grade', 'nova_group', 
                'additives_tags', 'sugars_100g', 'salt_100g']

df_interet = df[cols_interet]
```

```python

# 9. Afficher un aperçu des colonnes d'intérêt
display(df_interet.head())
```

```python
# Supprimer les doublons
df_clean = df.drop_duplicates(subset=['product_name', 'brands'])
```

```python
# Supprimer les lignes où toutes les colonnes d’intérêt sont NaN
df_clean = df_clean.dropna(subset=['nutriscore_grade', 'categories', 'brands'], how='all')
```

```python

# Remplacer les NaN par 'unknown' pour certaines colonnes
for col in ['brands', 'categories', 'nutriscore_grade']:
    df_clean[col] = df_clean[col].fillna('unknown')
```

```python
# Produits à Nutri-Score D/E ou NOVA 4
controverse = df_clean[
    (df_clean['nutriscore_grade'].isin(['d', 'e'])) | (df_clean['nova_group'] == 4)
]

# Top marques à risque
top_marques = controverse['brands'].value_counts().head(10)
print(top_marques)
```

```python
import matplotlib.pyplot as plt

# Répartition du Nutri-Score
df_clean['nutriscore_grade'].value_counts().sort_index().plot(
    kind='bar', color='skyblue')
plt.title("Répartition des Nutri-Score (A → E)")
plt.xlabel("Nutri-Score")
plt.ylabel("Nombre de produits")
plt.show()
```

```python

#Répartition des NOVA group
df_clean['nova_group'].value_counts().sort_index().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))
plt.title("Répartition des groupes NOVA")
plt.ylabel("")
plt.show()
```

```python
#Top 10 des marques les plus présentes
df_clean['brands'].value_counts().head(10).plot(
    kind='barh', color='orange')
plt.title("Top 10 des marques les plus présentes")
plt.xlabel("Nombre de produits")
plt.ylabel("Marque")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Distribution du taux de sucre par produit
df_clean['sugars_100g'].plot(
    kind='hist', bins=30, color='salmon', edgecolor='black')
plt.title("Distribution du taux de sucre (g/100g)")
plt.xlabel("Sucres (g/100g)")
plt.ylabel("Nombre de produits")
plt.show()
```

```python

# Scatter plot sucre vs sel

df_clean.plot.scatter(x='sugars_100g', y='salt_100g', alpha=0.5)
plt.title("Lien entre sucre et sel (g/100g)")
plt.xlabel("Sucres (g/100g)")
plt.ylabel("Sel (g/100g)")
plt.show()
```

```python
import seaborn as sns

# Sélection des variables numériques pertinentes
df_num = df_clean[['sugars_100g', 'salt_100g', 'fat_100g', 'proteins_100g']].dropna()

# Matrice de corrélation
corr = df_num.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Corrélation entre variables nutritionnelles")
plt.show()
```

```python
# Définir l’ordre voulu des catégories Nutri-Score
order_nutri = ['a', 'b', 'c', 'd', 'e', 'not-applicable', 'unknown']

plt.figure(figsize=(7,5))
sns.boxplot(
    x='nutriscore_grade',
    y='additives_n',
    data=df_clean,
    order=order_nutri
)
plt.title("Nombre d'additifs selon le Nutri-Score")
plt.xlabel("Nutri-Score (A = meilleur)")
plt.ylabel("Nombre d'additifs")
plt.show()
```

```python
#Top 10 Marques à risque (Nutri-Score D/E ou NOVA 4)

# Filtrer les produits à risque (Nutri-Score D/E ou NOVA 4)
produits_risque = df_clean[
    (df_clean['nutriscore_grade'].isin(['d', 'e'])) | (df_clean['nova_group'] == 4)
]

# Marques à risque
top_marques = (
    produits_risque['brands']
    .str.split(',', expand=True)
    .stack()
    .str.strip()
    .value_counts()
    .head(10)
)

print("Top 10 marques parmi les produits à risque :")

# Graphique
plt.figure(figsize=(7,4))
top_marques[::-1].plot(kind='barh', color='crimson')
plt.title("Top 10 marques (produits D/E ou NOVA 4)")
plt.xlabel("Nombre de produits à risque")
plt.ylabel("Marque")
plt.tight_layout()
plt.show()
```

```python
#Top 10 Catégories à risque (Nutri-Score D/E ou NOVA 4)

# Catégories à risque
top_categories = (
    produits_risque['categories']
    .str.split(',', expand=True)
    .stack()
    .str.strip()
    .value_counts()
    .head(10)
)

print("Top 10 catégories parmi les produits à risque :")

plt.figure(figsize=(7,4))
top_categories[::-1].plot(kind='barh', color='orange')
plt.title("Top 10 catégories (produits D/E ou NOVA 4)")
plt.xlabel("Nombre de produits à risque")
plt.ylabel("Catégorie")
plt.tight_layout()
plt.show()
```
